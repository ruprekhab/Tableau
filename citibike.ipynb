```python
import pandas as pd
from pathlib import Path
# import datetime as dt
```

```python
# Read the csv files

sept = pd.read_csv("Resources/sept.csv")
oct = pd.read_csv("Resources/oct.csv")
nov = pd.read_csv("Resources/nov.csv")
dec = pd.read_csv("Resources/dec.csv")

# Display  dec dataframe 
dec.head()
```

```python
# Concat july to dec dataframes
combined_df = pd.concat([sept, oct, nov, dec], ignore_index=True)
combined_df
```

```python
# Create a copy of combined_df
df = combined_df.copy()
```

```python
# Convert started_at and ended_at to date time datatype
df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])
```

```python
# Drop rows with missing values
df1 = df.dropna(how='any')
df1
```

```python
# New dataframe where date is more or equal to '2024-07-01" and less than equal to '2024-12-31'
df2 = df1[(df1['started_at'] >= "2024-09-01") & (df1['ended_at']<= '2024-12-31')]
df2
```

```python
# Save to csv
df2.to_csv("clean_data/data_0712.csv", index=False)
```
